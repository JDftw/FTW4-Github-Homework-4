```
quizdata = []

score = 0
questions = 10
quiz = dict()
print ("Please enter your name: ")
quiz['name'] = input().title()

print("1. What is the capital of the Republic of the Philippines? ")
print("a. Manila b. Davao c. Ilocos d. Batanes")
answer = input().lower()
answer = (answer)

if answer == 'a':
   print("correct")
   score = score + 1
else:
   print("wrong") 

print("2. What is the Summer Capital of the Philippines? ")
print("a. Manila b. Baguio City c. Rizal d. Tagaytay")
answer = input().lower()
answer = (answer)

if answer == 'b':
   print("correct")
   score = score + 1
else:
   print("wrong")

print("3. What is the oldest city of the Philippines? ")
print("a. Manila b. Davao c. Cebu d. Batanes")
answer = input().lower()
answer = (answer) 

if answer == 'c':
   print("correct")
   score = score + 1
else:
   print("wrong")

print("4. What is the largest island of the Philippine Archipelago? ")
print("a. Luzon b. Visayas c. Mindanao d. Batanes")
answer = input().lower()
answer = (answer)

if answer == 'a':
   print("correct")
   score = score + 1
else:
   print("wrong")

print("5. What is the longest mountain range in the Country? ")
print("a. Sierra Madre b. Mountain c. Seirra d. Madre")
answer = input().lower()
answer = (answer)

if answer == 'a':
   print("correct")
   score = score + 1
else:
   print("wrong")

print("6. Where is the highest peak in the Philippines? ")
print("a. Mt. Pulag b. Mt. Pinatubo c. Mt. Kanlaon d. Mt. Apo")
answer = input().lower()
answer = (answer)

if answer == 'd':
   print("correct")
   score = score + 1
else:
   print("wrong")

print("7. What is the longest river in the Philippines? ")
print("a. Marikina River b. Morong River c. Cagayan River d. Padsan River")
answer = input().lower()
answer = (answer)
  
if answer == 'c':
    print("correct")
    score = score + 1
else:
    print("wrong")

print("8. What is the provincial capital of Isabela? ")
print("a. Santiago b. Ilagan c. Cabagan d. Tumauini")
answer = input().lower()
answer = (answer)

if answer == 'b':
     print("correct")
     score = score + 1
else:
     print("wrong")

print("9. The Philippines is composed of approximately how many islands? ")
print("a. 7000 b. 7200 c. 7400 d. 7500")
answer = input().lower()
answer = (answer)

if answer == 'd':
    print("correct")
    score = score + 1
else:
    print("wrong")

print("10. What is the largest province in the Philippines in terms of land area? ")
print("a. Palawan b. Rizal c. Quezon d. Batangas")
answer = input().lower()
answer = (answer)

if answer == 'a':
     print("correct")
     score = score + 1
else:
     print("wrong")
     quiz['score'] = score
     quizdata.append(quiz)

score = round(score/questions*100)
print('Quiz completed.')
print('Your final score is {}%.'.format(score))

if int(score) >= 70:
    pass_fail = "Congrats, you passed!"
else:
    pass_fail = "Sorry, you failed."

print(f"{pass_fail}")
```

**Implement Create, Update and Delete in Quiz Q&A**

```
  quizfile = open("q&a.txt","r")
  filelist = quizfile.readlines()
  quizfile.close()    
  
  quizlist = []
  for k, line in enumerate(filelist):
    tuple_complete = False
    line1 = line.strip()

    if k % 6 == 0:
        your_question = line1
    if k % 6 == 1:
        letter_a = line1
    if k % 6 == 2:
        letter_b = line1
    if k % 6 == 3:
        letter_c = line1
    if k % 6 == 4:
        letter_d = line1
    if k % 6 == 5:
        correct_letter = line1
        tuple_complete = True
    if tuple_complete:
        tuple1 = (your_question, letter_a, letter_b, letter_c, letter_d, correct_letter)
        quizlist.append(tuple1)

    modify = 0

    while modify == 0:
  
      modify = input("Would you like to 'create', 'update' or 'delete' a question? \nType the word you want to process.\n")
      print(modify)

      #CREATE

    if modify == "create":
      your_question = input("\nThe question you want to add:\n") + "\n"
      letter_a = "a. " + input("First choice (a.): ") + "\n"
      letter_b = "b. " + input("Second choice (b.): ") + "\n"
      letter_c = "c. " + input("Third choice (c.): ") + "\n"
      letter_d = "d. " + input("Fourth choice (d.): ") + "\n"
      correct_letter = input("What is the correct answer? Just put the letter: ") + "\n"

      quizfile = open("q&a.txt","a")
      quizfile.write(your_question)
      quizfile.write(letter_a)
      quizfile.write(letter_b)
      quizfile.write(letter_c)
      quizfile.write(letter_d)
      quizfile.write(correct_letter)
      quizfile.close

      print ("Adding your question is successful.\n")
      modify = 0

      #UPDATE

    elif modify == "update":
        
      quizfile = open("q&a.txt","r")
      filelist = quizfile.readlines()
      quizfile.close()   

      print("\nThese are the questions on the quiz: ")
      for index, q in enumerate(filelist):
        if index % 6 == 0:
          print(f"[{index//6}] {q.strip()}")
      questionremove = input("\nPlease type the question you want to update: ")
      your_question = input("\nThe question you want to replace:\n") + "\n"
      change = open("q&a.txt").read().replace(questionremove, your_question)
    
   
      print ("The question was successfully updated.\n")
      modify = 0

      #DELETE

    elif modify == "delete":

      quizfile = open("q&a.txt","r")
      filelist = quizfile.readlines()
      quizfile.close()   

      print("\nThese are the questions on the quiz: ")
      for index, q in enumerate(filelist):
        if index % 6 == 0:
          print(f"[{index//6}] {q.strip()}")
      questionremove = input("\nWhich question would you like to remove? \nPlease enter the question number of your choice: ")
      qr = int(questionremove) * 6
      qr_range = [qr, qr+1, qr+2, qr+3, qr+4, qr+5, qr+6, qr+7, qr+8, qr+9]

      quizfile = open("q&a.txt","w")
      for index, q in enumerate(filelist):
        if index not in qr_range:
          quizfile.write(q)
      quizfile.close()    
      
      print ("The question was successfully removed.\n")
      modify = 0
```
